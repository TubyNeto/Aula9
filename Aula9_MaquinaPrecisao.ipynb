```python
import numpy as np

def calculaSolucao(a, b, n):
    s=0
    n=len(a)
    x=np.zeros(n)
    for k in range(n-1, -1, -1):
        s=0
        for j in range(k+1, n, 1):
            s=s+a[k][j]*x[j]
        x[k]=(b[k]-s)/a[k][k]            
    
    return x

def notacao(num, exp, lim):
  sinal = ''
  if (int(num.split('.')[0])>0 or (int(num.split('.')[0])<0) ) :

    integerSize=str(len(num.split('.')[0]))
    
    
    if(num[0] == '-'):
      sinal = num[0]
      print(sinal)
      num = num.replace(num[0],"")
      fractional = num.split('.')[0]

    elif(num[0]==0 and num[0] != '-'):

        fractional=num.split('.')[0] + num.split('.')[1] 
    else:

        fractional=num.split('.')[0] 
        

    if(len(fractional)>lim): fractional=fractional[:lim] 
    
    if(int(integerSize)>exp):

      frac = float(sinal + '0.'+fractional)
      new_exp = int(exp) 
      fractional='0.' + fractional + '*10^' + str(exp)
    else: 
      frac = float(sinal +'0.'+fractional)
      new_exp = int(integerSize) 
      fractional='0.' + fractional + '*10^' + integerSize

  else:

    fractional=num.split('.')[1] 

    i=0
    while (i!=len(fractional)):
      if (fractional!=None):  
          if (fractional[i]=='0'): i+=1
          else: break
    fractional=fractional.replace('0', '', i)

    if (len(fractional)>lim): fractional=fractional[:lim]

    if (i>=exp): 
      frac = float(sinal + '0.'+fractional)
      new_exp = int('-'+ str(exp)) 
      fractional='0.' + fractional + '*10^-' + str(exp)
    elif (i==0): 
      frac = float(sinal + '0.'+fractional)
      new_exp = 0 
      
      fractional='0.' + fractional + '*10^0'
    else: 
      frac = float(sinal + '0.'+fractional)
      str(i)
      new_exp = int('-'+str(i)) 
      fractional='0.' + fractional + '*10^-' + str(i)

  return [frac, new_exp]
  

def truncamento(num1, num2, t):

  exp1=(notacao(str(num1), 999, 3)[1])
  exp2=(notacao(str(num2), 999, 3)[1])

  num1=notacao(str(num1), 999, 3)[0]
  num2=notacao(str(num2), 999, 3)[0]

  size=0

  if (exp1 > exp2):
    num2 = str(num2)
    num2 = num2.split('.')[1]
    if (exp2 > 0):
      num2='0' * (exp1-exp2) + num2
    elif (exp2 == 0):
      num2='0' * exp1 + num2
    else:
      num2 = '0' * (exp1+exp2) + num2
    num2 = '0.' + num2
    
    size = len(num2.split('.')[1])
    if (size > t):
      return [0., num1*10**exp1]
    else:
      return [float(num2)*10**exp2, num1*10**exp1]



  else:
    num1 = str(num1)
    num1 = num1.split('.')[1]
    if (exp1 > 0):
      num1='0' * (exp2-exp1) + num1
    elif (exp1 == 0):
      num1='0' * exp2 + num1
    else:
      num1 = '0' * (exp2+exp1) + num1
    num1 = '0.' + num1
    
    size = len(num1.split('.')[1])
    if (size > t):
      return [num2*10**exp2, 0.]
    else:
      return [num2*10**exp2, float(num1)*10**exp1]




def pivoteamentoParcial(A, b, k, i, n):

  col = A[ i : n + 1, k ]

  colAbs = np.abs( col )

  indexMax = np.argmax( colAbs )


  iChange = indexMax + i

  aux = np.copy( A[ k, : ] )
  aux2 = 0

  A[ k, : ] = np.copy( A[ iChange, : ] )
  A[ iChange, : ] = np.copy( aux )
    
  aux2=b[k]
  b[k]=b[iChange]
  b[iChange]=aux2

  return A, b


def eliminacaoGaussTruncPivot( A, b, n, t ): 
    aux=0
    aux2=0
    for k in np.arange( 0, n ): 
        for i in np.arange( k + 1, n + 1 ):
            A, b = pivoteamentoParcial(A, b, k, i, n)
            m = A[ i, k ] / A[ k, k ]
            A[ i, k ] = 0
            for j in np.arange( k + 1, n + 1 ):
              aux = (m * A[ k, j ])
              A[ i, j ] = truncamento(aux, A[i][j], t)[0] - truncamento(aux, A[i][j], t)[1]
            aux2 = m * b[ k ] 
            b[ i ] = truncamento(aux2, b[i], t)[0] - truncamento(aux2, b[i], t)[1]
        
    return A, b  


def eliminacaoGaussTrunc( A, b, n, t ): 
    aux=0
    aux2=0
    for k in np.arange( 0, n ): 
        for i in np.arange( k + 1, n + 1 ):
            m = A[ i, k ] / A[ k, k ]
            A[ i, k ] = 0
            for j in np.arange( k + 1, n + 1 ):
              aux = (m * A[ k, j ])
              A[ i, j ] = truncamento(aux, A[i][j], t)[0] - truncamento(aux, A[i][j], t)[1]
            aux2 = m * b[ k ] 
            b[ i ] = truncamento(aux2, b[i], t)[0] - truncamento(aux2, b[i], t)[1]
        
    return A, b

def eliminacaoGaussPivot(A, b, n, t):
    for k in np.arange( 0, n ): 
        for i in np.arange( k + 1, n + 1 ): 
            A, b = pivoteamentoParcial(A, b, k, i, n)
            m = A[ i, k ] / A[ k, k ]
            A[ i, k ] = 0
            for j in np.arange( k + 1, n + 1 ):
                A[ i, j ] -= m * A[ k, j ]
            b[ i ] -= m * b[ k ] 

    return A, b


def eliminacaoGauss( A, b, n, t ): 
    for k in np.arange( 0, n ): 
        for i in np.arange( k + 1, n + 1 ): 
            m = A[ i, k ] / A[ k, k ]
            A[ i, k ] = 0
            for j in np.arange( k + 1, n + 1 ):
                A[ i, j ] -= m * A[ k, j ]
            b[ i ] -= m * b[ k ] 

    return A, b

def imprimeMatrizes(A, b, x, n, t):
  print("A ", A)
  print()
  print("b ", b)
  print()
  print("x ", x)
  print()


def resposta(A, b, t):
    
  Aarr=np.copy(A)
  barr=np.copy(b)
    
  Aarr2=np.copy(A)
  barr2=np.copy(b)
    
  Aarr3=np.copy(A)
  barr3=np.copy(b)
    
  print("\nUtilizando maquina de precisao\n")
  n = np.shape(A)[0] - 1

  print("Eliminacao gaussiana sem pivoteamento parcial")
  A1, b1 = eliminacaoGaussTrunc(A, b, n, t)
  x1 = calculaSolucao(A, b, n)
  imprimeMatrizes(A1, b1, x1, n, t)
    
  print("Eliminacao gaussiana com pivoteamento parcial")
  A3, b3 = eliminacaoGaussTruncPivot( Aarr2, barr2, n, t )
  x3 = calculaSolucao(A3, b3, n)
  imprimeMatrizes(A3, b3, x3, n, t)
    
    
  print("\nSem maquina de precisao\n")
  print("Eliminacao gaussiana sem pivoteamento parcial")
  A2, b2 = eliminacaoGauss(Aarr, barr, n, t)
  x2 = calculaSolucao(A2, b2, n) 
  imprimeMatrizes(A2, b2, x2, n, t)
    
  print("Eliminacao gaussiana com pivoteamento parcial")  
  A4, b4 = eliminacaoGaussPivot(Aarr3, barr3, n, t)
  x4 = calculaSolucao(A4, b4, n)
  imprimeMatrizes(A4, b4, x4, n, t)
    
  if (x1.all()==x2.all()):
    print('Resultado sem pivoteamento parcial igual com e sem máquina de precisão')
  else:
    print('Resultado sem pivoteamento parcial diferente com e sem máquina de precisão')
    
  if (x3.all()==x4.all()):
    print('Resultado com pivoteamento parcial igual com e sem máquina de precisão')
  else:
    print('Resultado com pivoteamento parcial diferente com e sem máquina de precisão')


A = np.array([[0.0002, 2.], [2., 2.]])
b = np.array([5., 6.])

A1=A.tolist()
b1=b.tolist()
n = np.shape(A)[0] - 1
t=3

for i in range (n+1):
    for j in range (n+1):
        A1[i][j] = str(notacao(str(A[i][j]), 9999, t)[0]) + '*10**' + str(notacao(str(A[i][j]), 9999, t)[1]) 

for k in range (n+1):
    b1[k] = notacao(str(b[k]), 9999, t)
    b1[k] = str(notacao(str(b[k]), 9999, t)[0]) + '*10**' + str(notacao(str(b[k]), 9999, t)[1])

print('A ', A1)
print('b ', b1)

resposta(A, b, t)
```
